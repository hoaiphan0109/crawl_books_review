```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
current_page = 1

data = []

proceed = True

while(proceed):
    print("Currently scraping page: "+str(current_page))
    
    url = "https://www.nxbtre.com.vn/diem-tin/review-sach/trang-"+str(current_page)+"/"

    page = requests.get(url)

    soup = BeautifulSoup(page.text, "html.parser")
    
    content_line = soup.find("span", {"id": "Content_ctl00_lb_empty"})
    
    if content_line.text.strip() == "Nội dung đang được cập nhật...":
        proceed = False
    else:
        all_books = soup.find_all("div",class_ = "col-md-3 col-sm-6") 
        
        for book in all_books:
            item = {}
            
            title_element = book.find("div", class_="title-diemtin")
            if title_element:
                item['Title'] = title_element.text.strip()
            else:
                item['Title'] = "N/A"
                
            href_element = book.find("a")
            if href_element:
                item['Link'] = "https://www.nxbtre.com.vn"+book.find("a").attrs["href"]
            else:
                item['Link'] = "N/A"
                        
            date_element = book.find("div", class_="datePost")
            if date_element:
                item['Date'] = date_element.text.split(":")[1].strip()
            else:
                item['Date'] = "N/A"
                
            if item['Title'] != "N/A": 
                data.append(item)
                
    current_page += 1

df = pd.DataFrame(data)
df
#df.to_excel("book_review.xlsx")

-- TEST CHANGE CODE

```
